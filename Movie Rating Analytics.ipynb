```python
#stacked Histogram thematic edit
sns.set_style('whitegrid')
list1=list()
mylabels=list()
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].BudgetMillions)
    mylabels.append(gen)

fig,ax=plt.subplots()
fig.set_size_inches(10.7,10.7)
h=plt.hist (list1, bins=30,stacked= True, rwidth=1 , label=mylabels)
plt.title("Movie Budget Distribution",fontsize=15,color='DarkBlue')
plt.ylabel('Number of Movies',fontsize=15,color='Black')
plt.xlabel('Budget',fontsize=15,color='Black')
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.legend(frameon=True, fancybox=True, shadow=True, framealpha=1,prop={'size':15})
plt.show()
```
